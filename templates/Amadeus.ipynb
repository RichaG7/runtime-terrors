```python
# Install the Python library from https://pypi.org/project/amadeus
# Get busiest months for each cities in 2017
amadeus = Client(
    client_id= api_key,
    client_secret= api_secret
)
cities = ['DEN','BOS','SFO','NYC','MSY']
for city in cities:
    try:
        response = amadeus.travel.analytics.air_traffic.busiest_period.get(cityCode=city, period='2017', direction='ARRIVING')
        print(f'We are looking at : {city}')
        jsonified_data = json.dumps(response.data, indent = 4, sort_keys = True)
        print(jsonified_data)
    except ResponseError as error:
        raise error
```
